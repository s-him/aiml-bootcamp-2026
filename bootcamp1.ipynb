```python
import pandas as pd
import numpy as np

df = pd.read_csv("session4_dummy_students_data.csv")
math_scores=df["math"].to_numpy()
eng_scores=df["english"].to_numpy()
sci_scores=df["science"].to_numpy()
df["avg"]=(math_scores+eng_scores+sci_scores)/3
df["result"] = np.where(df["avg"] >= 40, "pass", "fail")



print(df[["name","math","science","english","avg","result","attend"]])
```

```python
high_avg=df["avg"].max()
min_avg=df["avg"].min()
print(high_avg)
print(min_avg)
```

```python
low_attend = df[df['attendance_percent'] < 75]
print(low_attend[["name", "attendance_percent"]])
```
