```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv"
df = pd.read_csv(url)
continent_lifeexp = df.groupby("continent")["lifeExp"].mean().sort_values()
print(continent_lifeexp)
plt.figure(figsize=(8,5))
continent_lifeexp.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Average Life Expectancy by Continent")
plt.xlabel("Continent")
plt.ylabel("Average Life Expectancy")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv"
df = pd.read_csv(url)

# Filter only required countries
countries = ["India", "China", "United States"]
df_filtered = df[df["country"].isin(countries)]

# Pivot data: year as index, countries as columns
df_pivot = df_filtered.pivot(index="year", columns="country", values="lifeExp")

# Plot
plt.figure(figsize=(10,6))

plt.plot(df_pivot.index, df_pivot["India"], marker='o', label="India")
plt.plot(df_pivot.index, df_pivot["China"], marker='o', label="China")
plt.plot(df_pivot.index, df_pivot["United States"], marker='o', label="USA")

plt.title("Life Expectancy Over Time")
plt.xlabel("Year")
plt.ylabel("Life Expectancy")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv"
df = pd.read_csv(url)

plt.figure(figsize=(10, 6))

plt.scatter(
    df["gdpPercap"],
    df["lifeExp"],
    alpha=0.6
)

plt.xscale("log")

plt.title("GDP per Capita vs Life Expectancy")
plt.xlabel("GDP per Capita (Log Scale)")
plt.ylabel("Life Expectancy (Years)")
plt.grid(True)

plt.show()
```
